```python
"""
Knowledge Base Module
Manages primary (company) and secondary (industry) knowledge bases
"""

from src.document_processor import DocumentProcessor
from pathlib import Path


class KnowledgeBase:
    """Manages two-layer knowledge base system"""

    def __init__(self, primary_path, secondary_path):
        self.primary_path = Path(primary_path)
        self.secondary_path = Path(secondary_path)
        self.primary_processor = DocumentProcessor()
        self.secondary_processor = DocumentProcessor()
        self.is_loaded = False

    def load_all(self):
        """Load all documents from both knowledge bases"""
        print("Loading Primary Knowledge Base...")
        primary_docs = self.primary_processor.load_folder(self.primary_path)
        print(f"  Loaded {len(primary_docs)} primary documents")

        print("Loading Secondary Knowledge Base...")
        secondary_docs = self.secondary_processor.load_folder(self.secondary_path)
        print(f"  Loaded {len(secondary_docs)} secondary documents")

        self.is_loaded = True
        return {
            "primary_count": len(primary_docs),
            "secondary_count": len(secondary_docs)
        }

    def get_primary_context(self):
        """Get all primary knowledge base content"""
        return self.primary_processor.get_all_text()

    def get_secondary_context(self):
        """Get all secondary knowledge base content"""
        return self.secondary_processor.get_all_text()

    def get_full_context(self):
        """Get combined context from both knowledge bases"""
        return f"""
=== PRIMARY KNOWLEDGE BASE (Company-Specific) ===
{self.get_primary_context()}

=== SECONDARY KNOWLEDGE BASE (Industry Research) ===
{self.get_secondary_context()}
"""

    def search(self, keyword):
        """Search across both knowledge bases"""
        primary_results = self.primary_processor.search_documents(keyword)
        secondary_results = self.secondary_processor.search_documents(keyword)
        return {
            "primary": primary_results,
            "secondary": secondary_results
        }

    def get_summary(self):
        """Get summary of both knowledge bases"""
        return {
            "primary": self.primary_processor.get_document_summary(),
            "secondary": self.secondary_processor.get_document_summary()
        }
```
